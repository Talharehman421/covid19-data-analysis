```python
import pandas as pd
import numpy as np

df = pd.read_csv('data/raw/covid_data.csv')

# Always start with these checks
print(df.shape)
print(df.dtypes)
print(df.isnull().sum())
print(df.duplicated().sum())

# Fix date column
df['Date'] = pd.to_datetime(df['Date'])

# Fill missing numeric values with forward fill (common for time-series)
df[['Confirmed', 'Deaths', 'Recovered']] = df[['Confirmed', 'Deaths', 'Recovered']].fillna(method='ffill')

# Create derived columns — this shows analytical thinking
df['Active'] = df['Confirmed'] - df['Deaths'] - df['Recovered']
df['Fatality_Rate'] = (df['Deaths'] / df['Confirmed'] * 100).round(2)
df['Recovery_Rate'] = (df['Recovered'] / df['Confirmed'] * 100).round(2)

# Save cleaned data
df.to_csv('data/cleaned/covid_cleaned.csv', index=False)
print("Cleaning done! Shape:", df.shape)
```
